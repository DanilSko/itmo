```python
import re
```

```python
# напишем простой токенизатор
def simple_tokenizer (some_text):
    tokenized_text = re.sub (r'(?<![0-9]) (?![0-9])', ' || ', some_text)
    tokenized_text = re.sub (r'[.,:!](?= )', ' || ', tokenized_text)
    return tokenized_text
```
